```python
import pandas as pd
import os

# 경로 설정
input_csv = "../dataset/only_new_1204.csv"
output_csv = "../dataset/mask_channel_1204.csv"
mask_dir = "../dataset/mask_image_1204/"

# CSV 불러오기
df = pd.read_csv(input_csv)

# mask_image 열 초기화
df['mask_image'] = ""

# mask_dir 안의 이미지 파일 리스트 가져오기
# 필요하다면 확장자를 특정할 수 있음 예: if f.endswith('.jpg') or f.endswith('.png')
mask_images = os.listdir(mask_dir)

i = 0

# 각 행에 대해 image_path 열을 확인하고 mask_image를 채움
for idx, row in df.iterrows():
    image_path_value = row['image_path']
    for img_name in mask_images:
        i+=1
        if i % 100 == 0:
            print(img_name)
        # 이미지 이름이 image_path 문자열에 포함되어 있으면 해당 행에 mask_image 할당
        if img_name in image_path_value:
            df.at[idx, 'mask_image'] = 'dataset/mask_image_1204/' + img_name
            break  # 여러개가 있을 경우 첫 번째 매칭 후 중단

# 수정된 DataFrame을 새로운 CSV로 저장
df.to_csv(output_csv, index=False)
print("done")
```
