```python
import os
from PIL import Image

def convert_images_to_grayscale(relative_path):
    """
    Converts all images in a folder to grayscale and saves them to a new folder.
    
    Parameters:
    - input_dir (str): Path to the folder containing the original images.
    - output_dir (str): Path to the folder where grayscale images will be saved.
    """

    # Iterate over all files in the input directory
    for filename in os.listdir(relative_path):
        input_path = os.path.join(relative_path, filename)
        output_path = os.path.join(relative_path, filename)
        
        try:
            # Open the image and convert it to grayscale
            with Image.open(input_path) as img:
                grayscale_image = img.convert('L')  # Convert to grayscale
                grayscale_image.save(output_path)  # Save to the output directory
            print(f"Converted and saved: {output_path}")
        except Exception as e:
            print(f"Could not process file {filename}: {e}")

# Example usage
kde_graph_path = 'dataset/ku_kde_norm_1213/'
relative_path = '../' + kde_graph_path
convert_images_to_grayscale(relative_path)
```

```python
import pandas as pd

# 라벨과 등급에 따라 이미지 파일 매칭 함수 정의
def get_image_file(label, grade, kde_graph_path):
    # '등심' 제거 후 실제 등급 값 추출
    if grade.startswith("등심"):
        grade = grade.replace("등심", "")
    grade_mapping = {
        "1++": "1++",
        "1+": "1+",
        "1": "1",
        "2": "2",
        "3": "3"
    }
    if grade in grade_mapping:
        return f"{kde_graph_path}/kde_{label}_{grade_mapping[grade]}_grade.png"
    return None

# 사용자가 지정한 라벨
label = "Marbling"

# 경로에 있는 CSV 파일 읽기
csv_path = '../dataset/default_1213_fixed.csv'  # CSV 파일 경로
kde_graph_path = 'dataset/ku_kde_norm_1213'
data = pd.read_csv(csv_path)

# kde_graph 열 생성
data['kde_graph'] = data['grade'].apply(lambda x: get_image_file(label, x, kde_graph_path))


# 새 CSV 파일로 저장
output_csv_path = '../dataset/kde_marbling_1213.csv'  # 저장할 파일 경로
data.to_csv(output_csv_path, index=False)

print(f"새로운 CSV 파일이 생성되었습니다: {output_csv_path}")
```
